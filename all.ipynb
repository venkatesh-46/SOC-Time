```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
for f in os.listdir():
    f_name,f_ext=os.path.splitext(f)
    f_bin,f_n=f_name.split("_")
    df=pd.read_csv(f_bin+'_'+f_n+'.csv')
    
    z=df.SoC
    plt.hist(z,histtype='bar',bins=100)
    plt.xlabel('SoC in percent')
    plt.ylabel('Frequency')
    plt.title(f_name+' '+'All_sessions'+' '+'SoC in percent vs Frequency')
    plt.savefig(f_name+' '+'All_sessions'+' '+'histogram'+'.png')    
    plt.show()
   
    
    plt.plot('SoC','Time',data=df)
    plt.xlabel('SOC in percent')
    plt.ylabel('Time in mins')
    plt.title(f_name+' '+'All_sessions'+' '+'SoC in percent vs Time in mins')
    plt.savefig(f_name+' '+'All_sessions'+' '+'line_plot'+'.png')
    plt.show()
    
    
    plt.figure(figsize=(40,8))
    ax = sns.barplot(x='SoC', y='Time', data=df,orient = 'v')
    
    
    for index,row in df.iterrows():
        ax.text(row.name,row.Time,round(row.Time,2),color='black',ha="center")
    
    
    ax.set_ylabel('Time in mins')
    ax.set_xlabel('SOC in percent')
    plt.title(f_name+' '+'All_sessions'+' '+'SoC in percent vs Time in mins')
    plt.savefig(f_name+' '+'All_sessions'+' '+'Barplot'+'.png')
    plt.show()
    
    
    
```
